### Por qu es interesante estudiar la Hsp90? 
La protena de choque trmico de 90 kDa (Hsp90) es una protena chaperona responsable de catalizar la conversin de una amplia variedad de protenas a una forma funcional; ejemplos de la clientela de Hsp90, que asciende a varios cientos de protenas, incluyen receptores de hormonas esteroides nucleares y protena quinasas (Pearl y Prodromou 2006). El mecanismo por el cual acta Hsp90 vara entre los clientes, al igual que el sitio de enlace del cliente; el proceso depende de las modificaciones postraduccionales de Hsp90 y de la identidad de las cochaperonas que se unen y regulan el ciclo conformacional (Schopf et al. 2017). 

Debido a su papel bioqumico vital como protena chaperona involucrada en facilitar el plegamiento de muchas protenas cliente, Hsp90 es un objetivo farmacutico atractivo. En particular, dado que el plegamiento de protenas es un cuello de botella potencial para la reproduccin y el crecimiento celular, el bloqueo de la funcin Hsp90 mediante inhibidores que se unen estrechamente al sitio de unin de ATP del NTD podra ayudar en el tratamiento del cncer; por ejemplo, el antibitico geldanamicina y sus anlogos estn bajo investigacin como posibles agentes antitumorales (Stebbins et al. 1997, Hermane et al. 2019).
