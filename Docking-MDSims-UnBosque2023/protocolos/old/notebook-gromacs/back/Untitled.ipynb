# Análisis y dinámica molecular de alto rendimiento

## Introducción
Este tutorial proporciona una introducción al uso de la dinámica molecular de alto rendimiento para estudiar la interacción proteína-ligando, aplicada al dominio N-terminal de Hsp90 (proteína de choque térmico 90).

## Antecedentes
### ¿Qué es la dinámica molecular de alto rendimiento? 
La dinámica molecular (MD) es un método para simular el movimiento molecular mediante la aplicación iterativa de las leyes de movimiento de Newton. A menudo se aplica a biomoléculas grandes como proteínas o ácidos nucleicos. Una aplicación común es evaluar la interacción entre estas macromoléculas y varias moléculas pequeñas (p. ej., posibles candidatos a fármacos). Este tutorial proporciona una guía para configurar y ejecutar un flujo de trabajo de alto rendimiento para la detección de múltiples moléculas pequeñas, utilizando las herramientas GROMACS de código abierto proporcionadas a través de la plataforma Galaxy. Después de la simulación, los datos de la trayectoria se analizan utilizando una variedad de herramientas para investigar las propiedades estructurales y las correlaciones a lo largo del tiempo.

### ¿Por qué es interesante estudiar la Hsp90? 
La proteína de choque térmico de 90 kDa (Hsp90) es una proteína chaperona responsable de catalizar la conversión de una amplia variedad de proteínas a una forma funcional; ejemplos de la clientela de Hsp90, que asciende a varios cientos de proteínas, incluyen receptores de hormonas esteroides nucleares y proteína quinasas (Pearl y Prodromou 2006). El mecanismo por el cual actúa Hsp90 varía entre los clientes, al igual que el sitio de enlace del cliente; el proceso depende de las modificaciones postraduccionales de Hsp90 y de la identidad de las cochaperonas que se unen y regulan el ciclo conformacional (Schopf et al. 2017). 

Debido a su papel bioquímico vital como proteína chaperona involucrada en facilitar el plegamiento de muchas proteínas cliente, Hsp90 es un objetivo farmacéutico atractivo. En particular, dado que el plegamiento de proteínas es un cuello de botella potencial para la reproducción y el crecimiento celular, el bloqueo de la función Hsp90 mediante inhibidores que se unen estrechamente al sitio de unión de ATP del NTD podría ayudar en el tratamiento del cáncer; por ejemplo, el antibiótico geldanamicina y sus análogos están bajo investigación como posibles agentes antitumorales (Stebbins et al. 1997, Hermane et al. 2019). 

En la estructura que se examinará durante este tutorial, el ligando en cuestión es un resorcinol , una clase común de compuestos con afinidad por el dominio N-terminal de Hsp90. Está registrado en la base de datos de PubChem con el ID compuesto 135508238 (PubChem). Como se puede ver al observar la estructura PDB, la parte de resorcinol de la estructura está incrustada en el sitio de unión, unida por un enlace de hidrógeno al residuo aspartato-93. La estructura del ligando también contiene un triazol y un anillo de fluorofenilo, que se encuentran más cerca de la superficie de la proteína.


### Obtener datos 
Como primer paso, creamos un nuevo historial Galaxy y luego descargamos una estructura cristalina para la proteína Hsp90 del Protein Data Bank (PDB). La estructura se proporciona con el código de acceso 6HHR (Schuetz et al. 2018) y muestra Hsp90 en complejo con un ligando que pertenece a la clase de resorcinol.
